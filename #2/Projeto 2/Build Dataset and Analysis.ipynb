```python
dataset = pd.read_csv("dataset.csv")
jobs = [
    "Engineer",
    "Teacher",
    "Doctor",
    "Waiter",
    "Manager",
    "Accountant",
    "Lawyer"
]

jobs_income = [
    [40000,60000],
    [20000,35000],
    [60000,100000],
    [10000,20000],
    [40000,60000],
    [40000,60000],
    [30000,80000],
]

city = [
    "Viseu",
    "Coimbra",
    "Porto",
    "Aveiro",
    "Lisboa",
    "Braga",
    "Faro",
    "Castelo Branco",
]

signs = [
    "Aquarius",
    "Pisces",
    "Aries",
    "Taurus",
    "Gemini",
    "Cancer",
    "Leo",
    "Virgo",
    "Libra",
    "Scorpio",
    "Sagittarius",
    "Capricorn"
]

def return_income(job):
    incomes = jobs_income[jobs.index(job)]
    return random.randint(incomes[0], incomes[1])
    

def generate_date_of_birth(yo):
    yyyy = 2020-yo
    mm = random.randint(1, 12)
    dd = random.randint(1, 30)
    return "%02d-%02d-%4d" % (dd,mm,yyyy)

    
dataset["job"] = np.array([np.random.choice(jobs,1)[0] for i in range(dataset.shape[0])])
dataset["income_per_year"] = list(map(lambda x: return_income(x),dataset["job"]))
dataset["age"] = [random.randint(23, 60) for i in range(dataset.shape[0])]
dataset["number_kids"] = [random.randint(0, 4) for i in range(dataset.shape[0])]
dataset["city"] = [random.sample(city, 1)[0] for i in range(dataset.shape[0])]
dataset["signs"] = [random.sample(signs, 1)[0] for i in range(dataset.shape[0])]
dataset = dataset.drop(columns=["address"])

dataset.to_csv("dataset_complete.csv")
```

## Analysis

```python
def analysis(name_file):
    dataset = pd.read_csv(name_file)
    display(HTML('<h4>Analysis by gender</h4'))
    display(dataset.groupby("gender").mean()[["income_per_year","number_kids"]])
    display(HTML('<h4>Analysis by city</h4'))
    display(dataset.groupby("city").mean()[["income_per_year","number_kids"]])
    display(HTML('<h4>Analysis by age</h4'))
    display(dataset.groupby("age").mean()[["income_per_year","number_kids"]])
    display(HTML('<h4>Analysis by job</h4'))
    display(dataset.groupby("job").mean()[["income_per_year","number_kids"]])
    display(HTML('<h4>Analysis by signs</h4'))
    display(dataset.groupby("signs").mean()[["income_per_year","number_kids"]])
```

```python
display(HTML('<h1>Original Dataset</h1>'))
display(HTML('<h2>Analysis</h2>'))
analysis("dataset_final.csv")
display(HTML('<h3>Distinction and Separation</h3>'))
display(pd.read_csv("original_sep.csv"))
display(HTML('<h3>Risk</h3>'))
display(pd.read_csv("original_risk.csv"))
```

```python
one = pd.read_csv("out_1.csv")["gender"]
print("One -",len(one[one == "*"]))
suppr_one = len(one[one == "*"])
two = pd.read_csv("out_2.csv")["gender"]
print("Two -",len(two[two == "*"]))
suppr_two = len(two[two == "*"])
```

```python
one = [0,2,2,1,2]

two = [0,3,1,2,2]
max_ = [1,3,3,3,3]
qids = ["gender","city","job","age","signs"]
plt.figure(figsize=(10,7))
plt.fill_between(qids, max_, color="red",
                 alpha=0.6, label='Highest Generalization')
plt.fill_between(qids, two, color="blue",
                 alpha=0.9, label='Two')
plt.fill_between(qids, one, color="green",
                 alpha=0.9, label='One')

plt.xlabel("QIDs")
plt.ylabel("Level of generalization")
plt.legend()
plt.show()
```

```python
def precision(act,max_,size,size_supression):
    d = size * len(act)    
    n = sum([ act[i]/max_[i]*(size-size_supression) + size_supression for i in range(len(act))])
    
    return 1-n/d
```

```python
precision(one,max_,1000-suppr_one,suppr_one)
```

```python
precision(two,max_,1000-suppr_two,suppr_two)
```
